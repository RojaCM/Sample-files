```python
category_variables = ['surgery', 'age', 'temp_of_extremities','peripheral_pulse',
       'mucous_membrane', 'capillary_refill_time', 'pain', 'peristalsis',
       'abdominal_distention', 'nasogastric_tube', 'nasogastric_reflux', 'rectal_exam_feces', 'abdomen',
      'abdomo_appearance', 'surgical_lesion','cp_data']

for i in category_variables:
  df[i]=pd.get_dummies(df[i])
```

```python
df
```

```python
# label encoder for target variable 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['outcome']= le.fit_transform(df.outcome)
```

```python
df['rectal_temp']=df['rectal_temp'].fillna(df['rectal_temp'].mode()[0])
df['pulse']=df['pulse'].fillna(df['pulse'].mode()[0])
df['respiratory_rate']=df['respiratory_rate'].fillna(df['respiratory_rate'].mode()[0])
df['nasogastric_reflux_ph']=df['nasogastric_reflux_ph'].fillna(df['nasogastric_reflux_ph'].mode()[0])
df['packed_cell_volume']=df['packed_cell_volume'].fillna(df['packed_cell_volume'].mode()[0])
df['total_protein']=df['total_protein'].fillna(df['total_protein'].mode()[0])
df['abdomo_protein']=df['abdomo_protein'].fillna(df['abdomo_protein'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
x = df.drop('outcome',axis=1)
y=df.outcome
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
print(x_train.shape)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model1 = RandomForestClassifier()
```

```python
model1.fit(x_train,y_train)
```

```python
y_pred1=model1.predict(x_test)
```

```python
a1 = accuracy_score(y_pred1,y_test)
```
